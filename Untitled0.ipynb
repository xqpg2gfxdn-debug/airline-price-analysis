```python
import pandas as pd
url = "https://raw.githubusercontent.com/xqpg2gfxdn-debug/airline-price-analysis/refs/heads/main/Copy%20of%20flight_prices.csv%20-%20Sheet1.csv"
df = pd.read_csv(url)
df.head()
```

```python
import matplotlib.pyplot as plt

#Average price by class
avg = df.groupby('class')['price'].mean().sort_values()
avg.plot(kind='bar',title='Average Ticket Price by Class',ylabel='price (£)')
plt.show()
```

```python
import sqlite3
conn = sqlite3.connect(':memory:')
df.to_sql('flights', conn, index=False, if_exists='replace')

query = """
SELECT class, AVG(price) AS avg_price, COUNT(*) AS n_flights
FROM flights
GROUP BY class;
"""
pd.read_sql_query(query, conn)
```
